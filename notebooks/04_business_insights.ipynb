## Business Insights – Superstore Sales Analysis

Objective:
To analyze sales, profit, customer segments, and discount behavior
in order to identify growth opportunities and loss-making areas.


```python
import pandas as pd

df = pd.read_csv("../dataset/processed/cleaned_superstore_data.csv")
```

### Insight 1: Overall Performance
- The superstore generated strong overall sales with healthy order volume.
- However, profit growth does not always scale with sales, indicating cost or discount-related inefficiencies.


```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()

total_sales, total_profit, total_orders, total_customers
```

### Insight 2: Regional Analysis
- Certain regions generate high sales but comparatively lower profit.
- This indicates aggressive discounting or higher logistics costs in those regions.
- Recommendation: Review pricing and shipping strategy for low-margin regions.


```python
region_analysis = df.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales', ascending=False)
region_analysis
```

### Insight 3: Product Category Performance
- Technology products are the most profitable category.
- Some sub-categories consistently generate losses despite good sales.
- Recommendation: Reduce discounts or reconsider inventory for loss-making sub-categories.


```python
category_analysis = df.groupby('Category')[['Sales','Profit']].sum()
subcategory_analysis = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Profit')
subcategory_analysis.head()
```

### Insight 4: Customer Segments
- Consumer segment contributes the highest revenue.
- Corporate and Home Office segments show better profit stability.
- Recommendation: Design targeted offers based on segment profitability, not just volume.


```python
segment_analysis = df.groupby('Segment')[['Sales','Profit']].sum()
segment_analysis
```

### Insight 5: Discount vs Profit
- Higher discounts often lead to reduced or negative profit.
- Discounting does not always drive profitable growth.
- Recommendation: Implement discount caps and evaluate ROI before heavy promotions.


```python
discount_profit = df.groupby('Discount')['Profit'].mean()
discount_profit
```

### Insight 6: Sales Trends Over Time
- Sales show seasonal and yearly fluctuations.
- Peak sales periods align with year-end and festive seasons.
- Recommendation: Increase inventory and marketing spend during high-demand periods.


```python
yearly_sales = df.groupby('Order Year')['Sales'].sum()
yearly_sales
```

## Final Conclusion
- The superstore has strong revenue potential but faces profit leakage due to discounting.
- Focusing on high-margin products and regions can significantly improve profitability.
- Data-driven pricing and segment-based strategies are recommended for sustainable growth.
